```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)

df = pd.read_excel('data/raw_data.xlsx')
df.head()
```

## Replace unknown values in categorical columns

```python
# Keep relevant columns
cols_to_keep = [
    'INCIDENT_DATE', 'INCIDENT_YEAR', 'INCIDENT_MONTH', 'TIME', 'TIME_OF_DAY',
    'AIRPORT', 'STATE', 'LATITUDE', 'LONGITUDE', 'RUNWAY', 'OPID', 'OPERATOR',
    'AIRCRAFT', 'AC_CLASS', 'AC_MASS', 'NUM_ENGS', 'TYPE_ENG',
    'PHASE_OF_FLIGHT', 'SPECIES', 'NUM_SEEN', 'NUM_STRUCK', 'SIZE',
    'DAMAGE_LEVEL', 'NR_INJURIES', 'NR_FATALITIES', 'COST_REPAIRS',
    'COST_OTHER', 'COST_REPAIRS_INFL_ADJ', 'COST_OTHER_INFL_ADJ',
    'HEIGHT', 'SPEED', 'DISTANCE', 'WARNED'
]

filtered_df = df[cols_to_keep].copy()

# -----------------------------
# Fill missing categorical values
# -----------------------------
cat_cols = [
    'STATE', 'TIME_OF_DAY', 'RUNWAY', 'AC_CLASS', 'TYPE_ENG',
    'PHASE_OF_FLIGHT', 'SIZE', 'DAMAGE_LEVEL', 'NUM_STRUCK', 'NUM_SEEN',
    'WARNED'
]

filtered_df[cat_cols] = filtered_df[cat_cols].fillna('Unknown')

# -----------------------------
# Convert numeric columns
# -----------------------------
numeric_cols = [
    'LATITUDE', 'LONGITUDE', 'HEIGHT', 'SPEED', 'DISTANCE',
    'NR_INJURIES', 'NR_FATALITIES', 'COST_REPAIRS', 'COST_OTHER',
    'COST_REPAIRS_INFL_ADJ', 'COST_OTHER_INFL_ADJ'
]

filtered_df[numeric_cols] = filtered_df[numeric_cols].apply(
    pd.to_numeric,
    errors='coerce'
)

# -----------------------------
# Date and time features
# -----------------------------
filtered_df['INCIDENT_DATE'] = pd.to_datetime(
    filtered_df['INCIDENT_DATE'],
    format='%Y-%m-%d',
    errors='coerce'
)

time_dt = pd.to_datetime(
    filtered_df['TIME'],
    format='%H:%M',
    errors='coerce'
)

filtered_df['HOUR'] = time_dt.dt.hour
filtered_df['TIME'] = time_dt.dt.time

filtered_df['TIME_BUCKET'] = pd.cut(
    filtered_df['HOUR'],
    bins=[-1, 5, 11, 17, 23],
    labels=['Night', 'Morning', 'Afternoon', 'Evening']
)

# -----------------------------
# Binary indicators
# -----------------------------
filtered_df['HAS_DAMAGE'] = (filtered_df['DAMAGE_LEVEL'] != 'N').astype(int)
filtered_df['HAS_INJURY'] = (filtered_df['NR_INJURIES'] > 0).astype(int)
filtered_df['HAS_FATALITY'] = (filtered_df['NR_FATALITIES'] > 0).astype(int)

# -----------------------------
# Cost totals
# -----------------------------
filtered_df['TOTAL_COST'] = (
    filtered_df['COST_REPAIRS'].fillna(0) +
    filtered_df['COST_OTHER'].fillna(0)
)

filtered_df['TOTAL_COST_INFL_ADJ'] = (
    filtered_df['COST_REPAIRS_INFL_ADJ'].fillna(0) +
    filtered_df['COST_OTHER_INFL_ADJ'].fillna(0)
)

# -----------------------------
# Clean categorical mappings
# -----------------------------
filtered_df['WARNED_CLEAN'] = filtered_df['WARNED'].map({
    'Yes': 1,
    'No': 0,
    'Unknown': np.nan
})

engine_map = {
    'A': 'Piston',
    'B': 'Turbojet',
    'C': 'Turboprop',
    'D': 'Turbofan',
    'E': 'Glider',
    'F': 'Helicopter',
    'Y': 'Other',
    'Unknown': 'Unknown'
}

filtered_df['ENGINE_TYPE'] = filtered_df['TYPE_ENG'].map(engine_map)

damage_map = {
    'N': 'None',
    'M': 'Minor',
    'M?': 'Undetermined',
    'S': 'Substantial',
    'D': 'Destroyed',
    'Unknown': 'Unknown'
}

filtered_df['DAMAGE_CATEGORY'] = filtered_df['DAMAGE_LEVEL'].map(damage_map)

phase_map = {
    'Take-off Run': 'Takeoff',
    'Departure': 'Takeoff',
    'Climb': 'Climb',
    'En Route': 'Cruise',
    'Descent': 'Descent',
    'Approach': 'Approach',
    'Arrival': 'Approach',
    'Landing Roll': 'Landing',
    'Taxi': 'Ground',
    'Parked': 'Ground',
    'Local': 'Unknown',
    'Unknown': 'Unknown'
}

filtered_df['PHASE_GROUP'] = filtered_df['PHASE_OF_FLIGHT'].map(phase_map)

filtered_df['OPERATOR_TYPE'] = filtered_df['OPID'].replace({
    'PVT': 'Private',
    'BUS': 'Business',
    'GOV': 'Government',
    'MIL': 'Military'
})

# -----------------------------
# Drop redundant columns
# -----------------------------
filtered_df = filtered_df.drop(
    columns=['TYPE_ENG', 'DAMAGE_LEVEL', 'PHASE_OF_FLIGHT']
)
```
